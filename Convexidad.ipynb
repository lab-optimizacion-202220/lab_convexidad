# Primer punto

Evalué esta función sobre segmentos de recta entre dos puntos x1 = (x1, y1) y x2 = (x2, y2).

```python
def punto_1(x1,x2):
  # Se extraen los puntos de x1 y x2
  x_inicial = x1[0]
  x_final = x2[0]
  y_inicial = x1[1]
  y_final = x2[1]
  # Se crean los vectores de x y y
  vx = np.linspace(x_inicial,x_final,100)
  vy = np.linspace(y_inicial,y_final,100)
  # Se crea la linea f(x) y el vector sobre el cual sera graficado
  fx = np.linspace(0,0,100)
  graphx = np.linspace(1,2,len(vx))
  for i in range(0,len(vx)):
    fx[i] = 3*((1 - vx[i])**2) * math.exp(-(vx[i]**2)-(vy[i]+1)**2) - 10 * (vx[i]/5 - vx[i]**3 - vy[i]**5) * math.exp(-(vx[i]**2)-(vy[i]**2)) - 1/3 * math.exp(-(vx[i]+1)**2-(vy[i]**2))
  # Se encuentran los valores en f(x) de x1 y x2
  y1 = 3*((1 - x_inicial)**2) * math.exp(-(x_inicial**2)-(y_inicial+1)**2) - 10 * (x_inicial/5 - x_inicial**3 - y_inicial**5) * math.exp(-(x_inicial**2)-(y_inicial**2)) - 1/3 * math.exp(-(x_inicial+1)**2-(y_inicial**2))
  y2 = 3*((1 - x_final)**2) * math.exp(-(x_final**2)-(y_final+1)**2) - 10 * (x_final/5 - x_final**3 - y_final**5) * math.exp(-(x_final**2)-(y_final**2)) - 1/3 * math.exp(-(x_final+1)**2-(y_final**2))
  x = [1, 2]
  y = [y1, y2]
  plt.plot(x,y,color="red")
  plt.plot(graphx,fx)
  plt.show()
```

```python
# función
def gam(x):
    ans = 3*((1 - x[0])**2) * math.exp(-(x[0]**2)-(x[1]+1)**2) - 10 * (x[0]/5 - x[0]**3 -x[1] **5) * math.exp(-(x[0]**2)-(x[1]**2)) - 1/3 * math.exp(-(x[0]+1)**2-(x[1]**2))
    return ans 
```

```python
#gradienrte de la función en el punto a
grada = nd.Gradient(gam)([0, 1.6])
#gradienrte de la función en el punto b
gradb = nd.Gradient(gam)([0.2, -1.6])
#gradienrte de la función en el punto c
gradc = nd.Gradient(gam)([1, 1.2])
#gradienrte de la función en el punto d
gradd = nd.Gradient(gam)([0.3, 0.3])
ax=plt.contour(xcont,ycont,Z,10)
plt.colorbar(ax,shrink=0.5,aspect=5)
plt.title("Gradiente función gamma")
ax = plt.axes()
# Flecha con dirección de gradiente evaluada en punto a
ax.arrow(0, 1.6,grada[0],grada[1],color="r", head_width=0.2, head_length=0.2)
# Flecha con dirección de gradiente evaluada en punto b
ax.arrow(0.2, -1.6,gradb[0],gradb[1],color="b", head_width=0.2, head_length=0.2)
# Flecha con dirección de gradiente evaluada en punto c
ax.arrow(1, 1.2,gradc[0],gradc[1],color="g", head_width=0.2, head_length=0.2)
# Flecha con dirección de gradiente evaluada en punto d 
ax.arrow(0.3, 0.3,gradd[0],gradd[1],color="purple", head_width=0.2, head_length=0.2)
#grilla
plt.grid()
```

La gráfica anteriro muestra la curva de nivel y la dirección del gradiente evaluado en los 4 puntos dados. El gradiente en una función vectorial indica la dirección de máximo crecimiento, lo cual permite identificar si un punto dado en la función se encuentra cerca de un punto crítico. 
Para el punto a (rojo) el gradiente indica la dirección del máximo local más cercano, el cual está ubicado muy cerca al mismo punto. El punto b (azul) se encuentra muy cerca a un mínimo local, y su gradiente apunta a la dirección que se aleja más rápido del míniimo. El punto c (verde) está localizado cerca al máximo de la función y su gradiente está indicando la forma más rápida de llegar al máximo. Finalmente, el punto d (morado) está ubicado en un mínimo local y su gradiente está apuntando a la dirección de máximo crecimiento local en la función.

```python
# Se inicializan las variables encontradas en el literal a
coor1 = np.array([0, 1.6])
coor2 = np.array([-1.5, 0.3])
coor3 = np.array([0.4, 0.3])
coor4 = np.array([0, 1.25])
coor5 = np.array([-0.5, -0.6])
coor6 = np.array([0.25, -1.6])
# Se crea la lista que contendra los maximos y minimos
max = []
max.append(coor1)
max.append(coor4)
max.append(coor5)
min = []
min.append(coor2)
min.append(coor3)
min.append(coor4)
# Función para encontrar el valor de gamma segun la coordenada
def gam(x):
    ans = 3*((1 - x[0])**2) * math.exp(-(x[0]**2)-(x[1]+1)**2) - 10 * (x[0]/5 - x[0]**3 -x[1] **5) * math.exp(-(x[0]**2)-(x[1]**2)) - 1/3 * math.exp(-(x[0]+1)**2-(x[1]**2))
    return ans 
# Se recorren las listas de minimos y maximos locales para encontrar el global    
maxGlobal = 0
val = 0
for maxLocal in max:
  x = gam(maxLocal)
  if val < x:
    maxGlobal = maxLocal
    val = x

minGlobal = 1000000000000
val2 = 0
for minLocal in min:
  z = gam(minLocal)
  if val2 > z:
    minGlobal = minLocal
    val2 = z

print(maxGlobal)
print(minGlobal)
```

##### Como se puede observar el punto critico 1 es un maximo global mientras que el 4 y 5 son maximos locales. Por parte de los minimos, el punto critico 2 es el minimo global mientras que el 3 y el 6 son minimos locales.
